# Converting categorical variable into interger variables

```python
#importing sklearn preprocessing ,this library will help us in converting categorical varible into integers.
from sklearn import preprocessing
```

```python
label_encoding=preprocessing.LabelEncoder()
```

```python
df['Ship Mode']=label_encoding.fit_transform(df['Ship Mode'])
```

```python
#dropping country as the column 'country' has all values United states so it wont be helppul in bringing out data insights.
df.drop('Country',axis=1)
```

```python
#lets look into the unique values of category
df['Category'].unique()
```

```python
#frequency
df['Category'].value_counts()
```

```python
#subcategory unique values
df['Sub-Category'].nunique()
```

```python
#frequency
df['Sub-Category'].value_counts()
```

```python
#PLotting to see the relation betwwen category and sub category ,now through this plot you can clearly find out the category of
##subcategory
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=df,color='y')
plt.show()
```

```python
#pie chart plotting with subcategory
plt.figure(figsize=(12,10))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
#looking into profit and sales of the subcategory products together
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
```

```python
#get an insight about how sales is highest at west region and least at south
plt.figure(figsize=(16,8))
sns.countplot('Sub-Category',hue='Region',data=df)
plt.show()
```

```python
#lets find out the cost price
df['Cost']=df['Sales']-df['Profit']
```

```python
#profit percent
df['Profit %']=(df['Profit']/df['Cost'])*100
```

```python
#looking into costs
df['Cost'].head()
```

```python
df.sort_values(['Profit %','Sales'],ascending=False).groupby('Profit %').head(5)
```
